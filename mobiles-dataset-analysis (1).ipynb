```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def clean_numeric(column):
    return pd.to_numeric(column.str.replace(r'[^0-9.]', '', regex=True), errors='coerce')
```

```python
price_columns = [
    'Launched Price (Pakistan)', 
    'Launched Price (India)', 
    'Launched Price (China)', 
    'Launched Price (USA)', 
    'Launched Price (Dubai)'
]

for col in price_columns:
    Data[col] = clean_numeric(Data[col])
```

```python
Data['Mobile Weight'] = clean_numeric(Data['Mobile Weight'].str.replace('g', '', 
                                                                        ))
```

```python
Data['RAM'] = clean_numeric(Data['RAM'].str.replace('GB', '', regex=False))
```

```python
Data['Battery Capacity'] = clean_numeric(Data['Battery Capacity'].str.replace('mAh', '', regex=False))
Data['Screen Size'] = clean_numeric(Data['Screen Size'].str.replace(' inches', '', regex=False))
```

```python
Data.info()
```

```python
Data.describe()
```

```python
Data = Data[Data['Launched Price (USA)'] != 39622.000000]
```

```python
Data
```

```python
import warnings
warnings.filterwarnings("ignore")
```

<div style="background-color: #B7B1F2; padding: 25px; border-radius: 10px;">
  <h1 style="color: #333;"><strong><center>Distribution of mobile launches by year</center></strong></h1>
</div>

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Launched Year', data=Data, palette='viridis')
plt.title('Distribution of Mobile Launches by Year')
plt.xlabel('Launched Year')
plt.ylabel('Number of Models')

plt.show()
```

<div style="background-color: #B7B1F2; padding: 25px; border-radius: 10px;">
  <h1 style="color: #333;"><strong><center>Average prices comparison across different countries</center></strong></h1>
</div>

```python
avg_prices = Data[price_columns].mean().sort_values()
plt.figure(figsize=(10, 6))
avg_prices.plot(kind='barh', color='#2D336B')
plt.title('Average Mobile Prices Comparison by Country')
plt.xlabel('Average Price (in local currency)')

plt.show()
```

<div style="background-color: #B7B1F2; padding: 25px; border-radius: 10px;">
  <h1 style="color: #333;"><strong><center>Top companies by number of models launched</center></strong></h1>
</div>

```python
top_companies = Data['Company Name'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_companies.plot(kind='bar', color='#2D336B')
plt.title('Top Companies by Number of Models Launched')
plt.xlabel('Company Name')
plt.ylabel('Number of Models')

plt.show()
```

<div style="background-color: #B7B1F2; padding: 25px; border-radius: 10px;">
  <h1 style="color: #333;"><strong><center>Battery Capacity vs Price Analysis</center></strong></h1>
</div>

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Battery Capacity', y='Launched Price (USA)', data=Data, color='#2D336B')
plt.title('Battery Capacity vs Price (USA)')
plt.xlabel('Battery Capacity (mAh)')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.show()
```

<div style="background-color: #B7B1F2; padding: 25px; border-radius: 10px;">
  <h1 style="color: #333;"><strong><center>Price vs RAM Analysis</center></strong></h1>
</div>

```python
# Price vs. RAM Analysis
plt.figure(figsize=(12, 6))
sns.boxplot(x='RAM', y='Launched Price (India)', data=Data, palette='Set2')
plt.title('Price vs. RAM (India)')
plt.xlabel('RAM')
plt.ylabel('Price (INR)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

<div style="background-color: #B7B1F2; padding: 25px; border-radius: 10px;">
  <h1 style="color: #333;"><strong><center>Average Price by Top Companies</center></strong></h1>
</div>

```python
# Average Price by Top Companies
top_company_prices = Data[Data['Company Name'].isin(top_companies.index)]
avg_price_by_company = top_company_prices.groupby('Company Name')['Launched Price (India)'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_price_by_company.index, y=avg_price_by_company.values, palette='Spectral')
plt.title('Average Price by Top Companies (India)')
plt.xlabel('Company Name')
plt.ylabel('Average Price (INR)')
plt.xticks(rotation=45)

plt.show()
```

<div style="background-color: #B7B1F2; padding: 25px; border-radius: 10px;">
  <h1 style="color: #333;"><strong><center>Price Trends Over the Years</center></strong></h1>
</div>

```python
# Price Trends Over the Years
plt.figure(figsize=(12, 6))
avg_price_by_year = Data.groupby('Launched Year')['Launched Price (India)'].mean()
sns.lineplot(x=avg_price_by_year.index, y=avg_price_by_year.values, marker='o', color='#2D336B')
plt.title('Average Price Trends Over the Years (India)')
plt.xlabel('Launched Year')
plt.ylabel('Average Price (INR)')

plt.show()
```
